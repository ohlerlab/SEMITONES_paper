# Spatial mouse brain scRNA-seq
In this notebook we process the 10X sagital mouse brain spatial RNA data. We load the package using SeuratData conform the https://satijalab.org/seurat/v3.2/spatial_vignette.html vignette. The data can be found at https://support.10xgenomics.com/spatial-gene-expression/datasets.

## 1. Set up notebook

```r
library(Seurat)
library(SeuratData)
library(ggplot2)
library(patchwork)
library(dplyr)
library(Matrix)
```

## 2. Load data

```r
InstallData("stxBrain")
```

```r
ant1 <- LoadData("stxBrain", type="anterior1")
```

## 3. Normalize data

```r
ant1 <- SCTransform(ant1, assay="Spatial", variable.features.n=NULL,
                    verbose=FALSE)
```

## 4. Save normalized data

```r
norm <- ant1@assays[["SCT"]]@data  # get normalized data
norm <- as.data.frame(norm)

write.table(norm, "../data/processed/10X_spatial_musbrain_ant1_norm.txt",
            sep="\t")
```

```r
coordinates <- GetTissueCoordinates(ant1)
coordinates <- format(coordinates, decimal.mark=".")

write.table(coordinates, "../data/raw/10X_spatial_musbrain_ant1_coords.txt",
            sep="\t")
```
